```python
import pandas as pd
import numpy as np
```

```python
credits = pd.read_csv('credits.csv')
credits
```

```python
meta = pd.read_csv('movies_metadata.csv')
meta
```

```python
meta.columns
```

```python
meta['release_date'] = pd.to_datetime(meta['release_date'], errors = 'coerce')
```

```python
meta['year'] = meta['release_date'].dt.year
```

```python
meta['year'].value_counts().sort_index()
```

```python
# extract 2017 year data
new_meta = meta.loc[meta.year==2017,['genres','id','title','year']]
new_meta
```

```python
new_meta['id'] = new_meta['id'].astype(int)
```

```python
data = pd.merge(new_meta, credits,on='id')
```

```python
pd.set_option('display.max_colwidth', 75)
```

```python
data
```

```python
import ast

for i in ['genres','cast','crew']:
    data[i] = data[i].map(lambda x: ast.literal_eval(x))
```

```python
def make_genresList(x):
    gen = []
    st = " "
    for i in x:
        if i.get('name') == 'Science Fiction':
            scifi = 'Sci-Fi'
            gen.append(scifi)
        else:
            gen.append(i.get('name'))
    if gen == []:
        return np.NaN
    else:
        return (st.join(gen))
```

```python
data['genres_list'] = data['genres'].map(lambda x: make_genresList(x))
```

```python
data['genres_list']
```

```python
data['cast'].iloc[1]
```

```python
def get_actor1(x):
    casts = []
    for i in x:
        casts.append(i.get('name'))
    if casts == []:
        return np.NaN
    else:
        return (casts[0])
```

```python
data['actor_1_name'] = data['cast'].map(lambda x: get_actor1(x))
```

```python
data['actor_1_name']
```

```python
def get_actor2(x):
    casts = []
    for i in x:
        casts.append(i.get('name'))
    if casts == [] or len(casts)<=1:
        return np.NaN
    else:
        return (casts[1])
```

```python
data['actor_2_name'] = data['cast'].map(lambda x: get_actor2(x))
```

```python
data['actor_2_name']
```

```python
def get_actor3(x):
    casts = []
    for i in x:
        casts.append(i.get('name'))
    if casts == [] or len(casts)<=2:
        return np.NaN
    else:
        return (casts[2])
```

```python
data['actor_3_name'] = data['cast'].map(lambda x: get_actor3(x))
```

```python
data['actor_3_name']
```

```python
data['crew'].iloc[0]
```

```python
def get_directors(x):
    dt = []
    st = " "
    for i in x:
        if i.get('job') == 'Director':
            dt.append(i.get('name'))
    if dt == []:
        return np.NaN
    else:
        return (st.join(dt))
```

```python
data['director_name'] = data['crew'].map(lambda x: get_directors(x))
```

```python
data['director_name']
```
