```python
# select necessary columns
data = data.loc[:,['director_name','actor_1_name','actor_2_name','actor_3_name','genres','movie_title']]
data.head(6)
```

```python
# replace all columns null values with unknown
for col in data.columns:
    data[col] = data[col].replace(np.nan, 'unknown')
```

```python
# check any null values are there in our data
data.isna().any()
```

```python
# lower case the movie titles
data['movie_title'] = data['movie_title'].str.lower()
```

```python
data.movie_title[0]
```

```python
# remove the ending unknown characters from movie titles
data['movie_title'] = data['movie_title'].apply(lambda x : x[:-1])
```

```python
data.head(10)
```

```python
data['comb'] = data['actor_1_name'] + ' ' + data['actor_2_name'] + ' '+ data['actor_3_name'] + ' '+ data['director_name'] +' ' + data['genres']
```
